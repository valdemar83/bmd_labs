```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "r_normal1": np.random.randn(20),
    "r_gamma": np.random.gamma(shape=1, scale=5, size=20),
    "r_uniform": np.random.uniform(0,100,size=20),
    "r_gumbel":np.random.gumbel(20, 5, 20),
    "r_normal2":np.random.normal(loc=5,scale=2,size=20),
    "r_poisson": np.random.poisson(2, size=20)
})

df
```

```python
n = int(df.shape[0])
a=[0.394, 0.3917, 0.3894, 0.3872, 0.385, 0.383, 0.3808, 0.3789, 0.377, 0.3751]

def function(data):
    if n % 2 == 0:
        k = int(n/2)
    else:
        k = int((n-1)/2)
    
    mn = np.mean(data)
    nm2 = 0
    
    for i in range(n):
        nm2 = nm2 + (data[i] - mn)**2
        
    b = 0
    for i in range (k):
        b += a[i] * (data[n-i-1] - data[i])
        
    w = (b**2)/nm2
    return w
```

```python
function(df["r_normal1"])
```

```python
names = ["Variable", "Statistics", "Distribution"]
table = pd.DataFrame(columns = names)
table["Variable"] = df.keys()
table["Distribution"] = ['Normal' if 'norm' in i else 'Non-normal' for i in table['Variable']]
table
```

```python
for i in range(table.shape[0]):
    if 'norm' in table["Variable"][i]:
        table['Statistics'][i] = f'{df[table["Variable"][i]].mean()} {[df[table["Variable"][i]].std()]}'
for i in range (table.shape[0]):
    cur = table["Variable"][i]
    if 'norm' not in cur:
        table["Statistics"][i] =f'{df[cur].median()}[{df[cur].quantile(0.25)} {df[cur].quantile(0.75)}]'        
table
```

```python
import matplotlib.backends.backend_pdf
import matplotlib.pyplot as plt
fig = plt.figure()
ax=fig.add_subplot(111)

cell_text = []
for row in range(len(table)):
    cell_text.append(table.iloc[row])

ax.table(cellText=cell_text, colLabels=table.columns, loc='center')
ax.axis('off')

pdf = matplotlib.backends.backend_pdf.PdfPages("table.pdf")
pdf.savefig(fig)
pdf.close()
```
