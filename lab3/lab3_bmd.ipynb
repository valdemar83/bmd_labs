```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('car_crashes')
```

```python
tips.head()
```

```python
sns.distplot(tips['alcohol'])
```

```python
sns.jointplot(x='speeding', y='alcohol', data=tips)
```

```python
sns.pairplot(tips)
```

```python
sns.barplot(x='speeding', y='alcohol', data=tips)
```

```python
import numpy as np
```

```python
sns.barplot(x='total', y='speeding', data=tips, estimator=np.std)
```

```python
sns.boxplot(x = 'total', y='alcohol', data=tips, palette='rainbow')
```

```python
sns.stripplot(x='total', y='alcohol', data=tips)
```
